## Importing & Preprocessing

```python
import pandas as pd
# CountVectorizer (Helps to create bag of words)
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
# library to check accuracy
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
df = pd.read_csv('data.csv', encoding="ISO-8859-1") # encoding is used to make dataset usable which contain many special characters
```

```python
df.head()
```

```python
df.tail()
```

```python
# Split into train/test data using date
train = df[df['Date']<'20150101']
test = df[df['Date']> '20141231']
```

```python
# Removing punctuations
data = train.iloc[:,2:27]
data.replace("[^A-Za-z]"," ", regex=True, inplace= True) # This ("^a-zA-Z") implies all except small a to z and capital A to Z
```

```python
# Renaming the column names for ease of access
list1 = [i for i in range(25)]
new_Index=[str(i) for i in list1]
data.columns = new_Index
data.head(5)
```

```python
# Converting headlines to lower case
for index in new_Index:
    data[index]=data[index].str.lower()
data.head(1)
```

```python
# Combining the row
''.join(str(x) for x in data.iloc[1,0:25])
```

```python
# Combining all the columns of a row in a single paragraph through iterartion

#Creating a blank list of Headlines
headlines=[]

# Iterarting the dataset and store the paragraphs in the above list
for row in range(0,len(data.index)):
    headlines.append(''.join(str(x) for x in data.iloc[row,0:25]))
```

```python
headlines[4]
```

## Using CountVector

```python
# Implement bag of words
countvector = CountVectorizer(ngram_range=(2,2)) # (ngram_range take words as 2 by 2 i.e. in sentence The dog is on a mat it will match as (the dog), (dog is), (is on) and so on for all the probabilities)
traindataset = countvector.fit_transform(headlines)
```

```python
# implement Randomforest Classifier
randomclassifier = RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,train['Label'])
```

```python
# Preprocessing for test data
test_transform = []
for row in range(0,len(test.index)):
    test_transform.append(''.join(str(x) for x in test.iloc[row,2:27]))

test_dataset = countvector.transform(test_transform)
predictions = randomclassifier.predict(test_dataset)
```

```python
matrix = confusion_matrix(test['Label'], predictions)
print(matrix)
score = accuracy_score(test['Label'], predictions)
print(score)
report = classification_report(test['Label'], predictions)
print(report)
```

## Using TfidfVectorizer

```python
tfidfvector = TfidfVectorizer(ngram_range=(2,2)) # (ngram_range take words as 2 by 2 i.e. in sentence The dog is on a mat it will match as (the dog), (dog is), (is on) and so on for all the probabilities)
traindataset = tfidfvector.fit_transform(headlines)
```

```python
# implement Randomforest Classifier
randomclassifier = RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,train['Label'])
```

```python
# Preprocessing for test data
test_transform = []
for row in range(0,len(test.index)):
    test_transform.append(''.join(str(x) for x in test.iloc[row,2:27]))

test_dataset = tfidfvector.transform(test_transform)
predictions = randomclassifier.predict(test_dataset)
```

```python
matrix = confusion_matrix(test['Label'], predictions)
print(matrix)
score = accuracy_score(test['Label'], predictions)
print(score)
report = classification_report(test['Label'], predictions)
print(report)
```
